```python
energy_data_updated = "Updated_Energy_Consumption.csv"
energy_data_df = pd.read_csv(energy_data_updated)
energy_data_df.head(10)
```

```python
coal_change_over_time = energy_data_df.groupby(["year"])["coal_production"].mean()
coal_change_per_capita = energy_data_df.groupby(["year"])["coal_prod_per_capita"].mean()
coal_change_over_time.head(10)
```

```python
gas_change_over_time = energy_data_df.groupby(["year"])["gas_production"].mean()
gas_change_per_capita = energy_data_df.groupby(["year"])["gas_prod_per_capita"].mean()
gas_change_over_time.head(10)
```

```python
oil_change_over_time = energy_data_df.groupby(["year"])["oil_production"].mean()
oil_change_per_capita = energy_data_df.groupby(["year"])["oil_prod_per_capita"].mean()
oil_change_over_time.head(10)
```

```python
energy_per_capita = energy_data_df.groupby(["year"])["energy_per_capita"].mean()
energy_per_capita.tail(10)
```

```python
energy_prod_change_by_year = pd.DataFrame({
    "Energy per Capita": energy_per_capita,
    "Coal Production": coal_change_over_time,
    "Coal Production Per Capita": coal_change_per_capita,
    "Gas Production": gas_change_over_time,
    "Gas Production Per Capita": gas_change_per_capita,
    "Oil Production": oil_change_over_time,
    "Oil Production Per Capita": oil_change_per_capita,
})

energy_prod_change_by_year.head(10)
```

```python
energy_prod_change_by_year.to_csv('Energy_Production_Per_Year', index=False)
```

```python
solar_energy_per_capita = energy_data_df.groupby(["country"])["solar_energy_per_capita"].mean()
solar_energy_per_capita.head(10)
```

```python
wind_energy_per_capita = energy_data_df.groupby(["country"])["wind_energy_per_capita"].mean()
wind_energy_per_capita.head(10)
```

```python
nuclear_energy_per_capita = energy_data_df.groupby(["country"])["nuclear_energy_per_capita"].mean()
nuclear_energy_per_capita.head(10)
```

```python
low_carbon_per_capita = energy_data_df.groupby(["country"])['low_carbon_energy_per_capita'].mean()
low_carbon_per_capita.head(10)
```

```python
gdp_avg = energy_data_df.groupby(['country'])['gdp'].mean()
gdp_avg.head(10)
```

```python
green_energy_per_country = pd.DataFrame({
    "Renewable Electricity": renew_elec_per_country,
    "Renewable Energy": renew_energy_per_capita,
    "Solar Energy": solar_energy_per_capita,
    "Coal Production Per Capita": coal_change_per_capita,
    "Gas Production": gas_change_over_time,
    "Gas Production Per Capita": gas_change_per_capita,
    "Oil Production": oil_change_over_time,
    "Oil Production Per Capita": oil_change_per_capita,
})

green_energy_per_country.head(10)
```
