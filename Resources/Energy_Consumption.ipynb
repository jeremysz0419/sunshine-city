```python
energy_data_updated = "Updated_Energy_Consumption.csv"
energy_data_df = pd.read_csv(energy_data_updated)
energy_data_df.head(10)
```

```python
coal_change_over_time = energy_data_df.groupby(["year"])["coal_consumption"].mean()
coal_change_per_capita = energy_data_df.groupby(["year"])["coal_cons_per_capita"].mean()
coal_change_over_time.head(10)
```

```python
gas_change_over_time = energy_data_df.groupby(["year"])["gas_consumption"].mean()
gas_change_elec_per_capita = energy_data_df.groupby(["year"])["gas_elec_per_capita"].mean()
gas_change_energy_per_capita = energy_data_df.groupby(["year"])["gas_energy_per_capita"].mean()
gas_change_over_time.head(10)
```

```python
oil_change_over_time = energy_data_df.groupby(["year"])["oil_production"].mean()
oil_change_elec_per_capita = energy_data_df.groupby(["year"])["oil_elec_per_capita"].mean()
oil_change_energy_per_capita = energy_data_df.groupby(["year"])["oil_energy_per_capita"].mean()
oil_change_over_time.head(10)
```

```python
energy_per_capita = energy_data_df.groupby(["year"])["energy_per_capita"].mean()
energy_per_capita.tail(10)
```

```python
energy_consump_change_by_year = pd.DataFrame({
    "Energy per Capita": energy_per_capita,
    "Coal Consumption": coal_change_over_time,
    "Coal Consumption Per Capita": coal_change_per_capita,
    "Gas Consumption": gas_change_over_time,
    "Gas Consumption Electricity Per Capita": gas_change_elec_per_capita,
    "Gas Consumption Energy Per Capita": gas_change_energy_per_capita,
    "Oil Consumption": oil_change_over_time,
    "Oil Consumption ELectricity Per Capita": oil_change_elec_per_capita,
    "Oil Consumption Energy Per Capita": oil_change_energy_per_capita,
})

energy_consump_change_by_year.head(10)
```

```python
energy_consump_change_by_year.to_csv('Energy_Consumption_Per_Year', index=False)
```

```python
nuclear_consumption = energy_data_df.groupby(["country"])["nuclear_consumption"].mean()
nuclear_elec_per_capita = energy_data_df.groupby(["country"])["nuclear_elec_per_capita"].mean()
nuclear_energy_per_capita = energy_data_df.groupby(["country"])["nuclear_energy_per_capita"].mean()
nuclear_consumption.head(10)
```

```python
hydro_consumption = energy_data_df.groupby(["country"])["hydro_consumption"].mean()
hydro_elec_per_capita = energy_data_df.groupby(["country"])["hydro_elec_per_capita"].mean()
hydro_energy_per_capita = energy_data_df.groupby(["country"])["hydro_energy_per_capita"].mean()
hydro_consumption.head(10)
```

```python
solar_consumption = energy_data_df.groupby(["country"])["solar_consumption"].mean()
solar_elec_per_capita = energy_data_df.groupby(["country"])["solar_elec_per_capita"].mean()
solar_energy_per_capita = energy_data_df.groupby(["country"])["solar_energy_per_capita"].mean()
solar_consumption.head(10)
```

```python
wind_consumption = energy_data_df.groupby(["country"])["wind_consumption"].mean()
wind_elec_per_capita = energy_data_df.groupby(["country"])["wind_elec_per_capita"].mean()
wind_energy_per_capita = energy_data_df.groupby(["country"])["wind_energy_per_capita"].mean()
wind_consumption.head(10)
```

```python
low_carbon_consumption = energy_data_df.groupby(["country"])["low_carbon_consumption"].mean()
low_carbon_per_capita = energy_data_df.groupby(["country"])["low_carbon_energy_per_capita"].mean()
low_carbon_consumption.head(10)
```

```python
renew_elec_consump_per_country = energy_data_df.groupby(["country"])["renewables_consumption"].mean()
renew_elec_consump_per_country_capita = energy_data_df.groupby(["country"])["renewables_energy_per_capita"].mean()
renew_elec_consump_per_country.head(10)
renew_elec_consump_per_country_capita.head(10)
```

```python
other_renewable_consump_per_country = energy_data_df.groupby(["country"])["renewables_consumption"].mean()
other_renewable_consump_per_country_capita = energy_data_df.groupby(["country"])["renewables_energy_per_capita"].mean()
other_renewable_consump_per_country.head(10)
other_renewable_consump_per_country_capita.head(10)
```

```python
gdp_avg = energy_data_df.groupby(['country'])['gdp'].mean()
gdp_avg.head(10)
```

```python
green_energy_per_country = pd.DataFrame({
    "Solar Energy Consumption": solar_consumption,
    "Solar Energy Electricity per Capita": solar_elec_per_capita,
    "Solar Energy per Capita": solar_energy_per_capita,
    "Wind Energy Consumption": wind_consumption,
    "Wind Energy Electricity per Capita": wind_elec_per_capita,
    "Wind Energy per Capita": wind_energy_per_capita,
    "Hydro Energy Consumption": hydro_consumption,
    "Hydro Energy Electricity per Capita": hydro_elec_per_capita,
    "Hydro Energy per Capita": hydro_energy_per_capita,
    "Nuclear Energy Consumption": nuclear_consumption,
    "Nuclear Energy Electricity per Capita": nuclear_elec_per_capita,
    "Nuclear Energy per Capita": nuclear_energy_per_capita,
    "Low Carbon Energy Consumption": low_carbon_consumption,
    "Low Carbon Energy per Capita": low_carbon_per_capita,
    "Renewable Energy Consumption": renew_elec_consump_per_country,
    "Renewable Energy per Capita": renew_elec_consump_per_country_capita,
    "Other Renewable Energy Consumption": other_renewable_consump_per_country,
    "Other Renewable Energy per Capita": other_renewable_consump_per_country_capita,
    "GDP": gdp_avg,
})

green_energy_per_country.head(10)
```

```python
green_energy_per_country.to_csv('Green_Energy_Per_Country', index=False)
```
